```python
duration_series = []
for x in df["Duration"]:
  #print(x)
  #print(type(x))
  h = 0
  m = 0
  l = x.split(" ")
  for y in l:
    if "h" in y:
      h += int(''.join(map(str, (y[:-1]))))

    if "m" in y:
      # print(type(y[:-1]))
      # print(y[:-1])
      m += int(''.join(map(str, (y[:-1]))))
      #y += int
  seconds = h*60*60 + m*60
  duration_series.append(seconds)
```

```python
df["Duration(seconds)"] = pd.Series(duration_series)
```
